```python
def agua_restante(astronautas, agua_restante, dias_restantes):
    uso_diario = astronautas * 11
    uso_total = uso_diario * dias_restantes
    agua_restante_total = agua_restante - uso_total
    if agua_restante_total < 0:
        raise RuntimeError(f"No hay suficiente agua para {astronautas} astronautas despu√©s de {dias_restantes} d√≠as!")
    return f"El agua restante total despu√©s de {dias_restantes} d√≠as es: {agua_restante_total} litros"
```

```python
def agua_restante(astronautas, agua_restante, dias_restantes):
    for argument in [astronautas, agua_restante, dias_restantes]:
        try:
            # Si el argumento es un entero, la siguiente operaci√≥n funcionar√°
            argument / 10
        except TypeError:
            # TypError saldr√° s√≥lo si no es el tipo correcto 
            # Misma excepci√≥n pero con un mejor mensje de error
            raise TypeError(f"Todos los argumentos deben ser del tipo int, pero recibi√≥: '{argument}'")
    uso_diario = astronautas * 11
    uso_total = uso_diario * dias_restantes
    agua_restante_total = agua_restante - uso_total
    if agua_restante_total < 0:
        raise RuntimeError(f"No hay suficiente agua para {astronautas} astronautas despu√©s de {dias_restantes} d√≠as!")
    return f"El agua restante total despu√©s de {dias_restantes} d√≠as es: {agua_restante_total} litros"
```
